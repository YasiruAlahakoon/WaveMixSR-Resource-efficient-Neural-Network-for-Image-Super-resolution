```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the Mexican Hat wavelet (Ricker wavelet) function
def mexican_hat_wavelet(x, scale):
    return (2 / (np.sqrt(3 * scale) * np.pi**0.25)) * (1 - (x / scale)**2) * np.exp(-(x**2) / (2 * scale**2))

# Generate x values
x = np.linspace(-5, 5, 400)

# Define different scaling factors
scales = [0.5, 1, 2, 4]

# Plot the Mexican Hat wavelet for each scale
plt.figure(figsize=(10, 6))

for scale in scales:
    y = mexican_hat_wavelet(x, scale)
    plt.plot(x, y, label=f'Scale = {scale}')

# Customize the plot
plt.title('Mexican Hat Wavelet for Different Scales')
plt.xlabel('x')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def belkin_wavelet(x, scale):
    return np.sinc(x / scale) * np.cos(2 * np.pi * x / scale)

# Generate x values
x = np.linspace(-5, 5, 400)

# Define different scaling factors
scales = [0.5, 1, 2, 4]

# Plot the Belkin wavelet for each scale
plt.figure(figsize=(10, 6))

for scale in scales:
    y = belkin_wavelet(x, scale)
    plt.plot(x, y, label=f'Scale = {scale}')

# Customize the plot
plt.title('Belkin Wavelet for Different Scales')
plt.xlabel('x')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def meyer_wavelet(x, scale):
    return np.sin(2 * np.pi * x / scale) * np.exp(-x**2 / (2 * scale**2))


# Generate x values
x = np.linspace(-5, 5, 400)

# Define different scaling factors
scales = [0.5, 1, 2, 4]

# Plot the Meyer wavelet for each scale
plt.figure(figsize=(10, 6))

for scale in scales:
    y = meyer_wavelet(x, scale)
    plt.plot(x, y, label=f'Scale = {scale}')

# Customize the plot
plt.title('Meyer Wavelet for Different Scales')
plt.xlabel('x')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def symlet_wavelet(x, scale):
    return (x**2 - 1) * np.exp(-x**2 / (2 * scale**2))


# Generate x values
x = np.linspace(-5, 5, 400)

# Define different scaling factors
scales = [0.5, 1, 2, 4]

# Plot the Mexican Hat wavelet for each scale
plt.figure(figsize=(10, 6))

for scale in scales:
    y = symlet_wavelet(x, scale)
    plt.plot(x, y, label=f'Scale = {scale}')

# Customize the plot
plt.title('Symlet Wavelet for Different Scales')
plt.xlabel('x')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```
